```python
# -*- coding: utf-8 -*-
"""
Created on Fri Aug 03 20:27:35 2018

@author: azmi
"""
import math
from gurobipy import *
try:
    # Create a new model
    
    m = Model("Dial-a-ride Problem")
    node =3  # Total request
    vehicle = 2 
    tnode = 4*node + 2*vehicle
    N = 4*node
    A = tnode
    
    #Declaring Variables and parameters with index
    #actually this phase only need to declare variable. but im too lazy to remove it
    X=[[[0 for k in range(A)] for j in range(A)] for i in range(A)]
    T=[[0 for x in range(A)] for y in range(A)]
    L=[[0 for x in range(A)] for y in range(A)]
    w=[0 for x in range(A)]
    dplus=[0 for x in range(A)]
    dmin=[0 for x in range(A)]
    c=[[0 for x in range(A)] for y in range(A)]
    t=[[0 for x in range(A)] for y in range(A)]
    a=[0 for x in range(A)]
    b=[0 for x in range(A)]
    C=[0 for x in range(vehicle)]
    l=[0 for x in range(A)]
    d=[0 for x in range(A)]
    PB=[0 for x in range(A)]
    DB=[0 for x in range(A)]
    PA=[0 for x in range(A)]
    DA=[0 for x in range(A)]
    OK=[0 for x in range(A)]
    DK=[0 for x in range(A)]
    
    #Define subset for index
    
    for i in range(node):
        PB[i] = i
    
    
    for i in range(node):
        DB[i] = node+i
    
    
    for i in range(node):
        PA[i] = 2*node+i
    
    for i in range(node):
        DA[i] = 3*node+i
          
    for k in range(vehicle):
        OK[k] = 4*node+k
    
    for k in range(vehicle):
        DK[k] = 4*node+vehicle+k
    
    
    
    
    #Declaring Parameters
    t[0] = [0,	0,	0.22,	0.6,	0.9,	0.6,	0.6,	0.9,	0.6,	0,	0,	0.22,	0,	0,	0,	0]
    t[1] = [0,	0,	0.22,	0.6,	0.9,	0.6,	0.6,	0.9,	0.6,	0,	0,	0.22,	0,	0,	0,	0]
    t[2] = [0.22,	0.22,	0,	0.7,	1.05,	0.7,	0.7,	1.05,	0.7,	0.22,	0.22,	0,	0.22,0.22,0.22,0.22]
    t[3] = [0.6,	0.6,	0.7,	0,	1.4,	0,	0,	1.4,	0,	0.6,	0.6,	0.7,	0.6,	0.6,	0.6,	0.6]
    t[4] = [0.9,	0.9,	1.05,	1.4,	0,	1.4,	1.4,	0,	1.4,	0.9,	0.9,	1.05,	0.9,	0.9,	0.9,	0.9]
    t[5] = [0.6,	0.6,	0.7,	0,	1.4,	0,	0,	1.4,	0,	0.6,	0.6,	0.7,	0.6,	0.6,	0.6,	0.6]
    t[6] = [0.6,	0.6,	0.7,	0,	1.4,	0,	0,	1.4,	0,	0.6,	0.6,	0.7,	0.6,	0.6,	0.6,	0.6]
    t[7] = [0.9,	0.9,	1.05,	1.4,	0,	1.4,	1.4,	0,	1.4,	0.9,	0.9,	1.05,	0.9,	0.9,	0.9,	0.9]
    t[8] = [0.6,	0.6,	0.7,	0,	1.4,	0,	0,	1.4,	0,	0.6,	0.6,	0.7,	0.6,	0.6,	0.6,	0.6]
    t[9] = [0,	0,	0.22,	0.6,	0.9,	0.6,	0.6,	0.9,	0.6,	0,	0,	0.22,	0,	0,	0,	0]
    t[10] = [0,	0,	0.22,	0.6,	0.9,	0.6,	0.6,	0.9,	0.6,	0,	0,	0.22,	0,	0,	0,	0]
    t[11] = [0.22,	0.22,	0,	0.7,	1.05,	0.7,	0.7,	1.05,	0.7,	0.22,	0.22,	0,	0.22,0.22,0.22,0.22]
    t[12] = [0,	0,	0.22,	0.6,	0.9,	0.6,	0.6,	0.9,	0.6,	0,	0,	0.22,	0,	0,	0,	0]
    t[13] = [0,	0,	0.22,	0.6,	0.9,	0.6,	0.6,	0.9,	0.6,	0,	0,	0.22,	0,	0,	0,	0]
    t[14] = [0,	0,	0.22,	0.6,	0.9,	0.6,	0.6,	0.9,	0.6,	0,	0,	0.22,	0,	0,	0,	0]
    t[15] = [0,	0,	0.22,	0.6,	0.9,	0.6,	0.6,	0.9,	0.6,	0,	0,	0.22,	0,	0,	0,	0]
    
    
    for i in range(A):
        for j in range(A):
            c[i][j] = t[i][j]
             
    #Parameter a&b
    a[3] = 12  
    b[3] = 12.5
    
    a[4] = 8.5 
    b[4] = 9
    
    a[5] = 12.5 
    b[5] = 13
     
    a[6] = 13.5 
    b[6] = 14
    
    a[7] = 13 
    b[7] = 13.5
     
    a[8] = 14 
    b[8] = 14.5
        
    #Parameter  C
    for k in range(vehicle):
        C[k] = 8   
    d[0] = 1
    d[1] = 4
    d[2] = 1
    
    d[6] = 1
    d[7] = 4
    d[8] = 1
    
    #Parameter d&l
    for i in range(node):
        
        l[PB[i]] = d[PB[i]]
        l[PA[i]] = d[PA[i]]
        l[node+PA[i]] = -d[PA[i]] 
        l[node+PB[i]] = -d[PB[i]]
        
    #Parameter alpha&beta
    beta = [1,1,1,1,1]
    
    alpha1 = 0.4
    alpha2 = 1-alpha1 
    
    # Create variables
    
    #define Xijk
    for i in range(A):
        for j in range(A):
            for k in range(vehicle):
                X[i][j][k] = m.addVar(vtype=GRB.BINARY, name="X["+str(i+1)+","+str(j+1)+","+str(k+1)+"]")
    #define Tik           
    for i in range(A):
        for k in range(vehicle):
             T[i][k]=m.addVar(lb = 0, vtype=GRB.CONTINUOUS, name="T["+str(i+1)+","+str(k+1)+"]")
             
    #define Lik
    for i in range(A):
        for k in range(vehicle):
            L[i][k]=m.addVar(lb = 0, vtype=GRB.INTEGER, name="L["+str(i+1)+","+str(k+1)+"]")
    
    #define wi
    for i in range(node):
        w[PA[i]]=m.addVar(vtype=GRB.CONTINUOUS, name="w["+str(PA[i]+1)+"]")
        
    #define dplus_i     
    for i in range(node):
        dplus[DB[i]]=m.addVar(vtype=GRB.CONTINUOUS, name="d+["+str(DB[i]+1)+"]")
    for i in range(node):
        dplus[PA[i]]=m.addVar(vtype=GRB.CONTINUOUS, name="d+["+str(PA[i]+1)+"]")
    
    #define dmin_i
    for i in range(node):
        dmin[DB[i]]=m.addVar(vtype=GRB.CONTINUOUS, name="d-["+str(DB[i]+1)+"]")

    totalcost = m.addVar(vtype=GRB.CONTINUOUS, name="totalcost")
    m.update()

    # Set objective (1)
    m.setObjective(alpha1*(quicksum(quicksum(quicksum(c[i][j]*X[i][j][k] for j in range(4*node+ 2*vehicle))for i in range(4*node+ 2*vehicle))for k in range(vehicle)) + beta[0]*quicksum(w[2*node+i] for i in range(node))) + alpha2*(quicksum(quicksum(beta[1]*((T[node+i][k] -T[i][k] -t[i][node+i])+(T[3*node+i][k] - T[2*node+i][k]-t[2*node+i][3*node+i]))+beta[2]*dplus[2*node+i]+beta[3]*dmin[node+i]+beta[4]*dplus[node+i] for i in range(node))for k in range(vehicle))), GRB.MINIMIZE )
    
    m.addConstr(quicksum(quicksum(X[4*node+k][PB[j]][k] for j in range(node)) for k in range(vehicle)) >= 1)
    
    #constrain (3)
    for i in range(node):
        m.addConstr(quicksum(quicksum(X[PB[i]][j][k] for j in range(N))for k in range(vehicle)) == 1)
        m.addConstr(quicksum(quicksum(X[PA[i]][j][k] for j in range(N))for k in range(vehicle)) == 1)
        
    #constraint (4)
    for i in range(node):
        for k in range(vehicle):
            m.addConstr(quicksum(X[PB[i]][j][k] for j in range(N) ) - quicksum(X[j][node+PB[i]][k] for j in range(N)) == 0 )
            m.addConstr(quicksum(X[PA[i]][j][k] for j in range(N) ) - quicksum(X[j][node+PA[i]][k] for j in range(N)) == 0 )
    
    #constraint (5)
    for k in range(vehicle):
        m.addConstr(quicksum(X[(4*node)+k][PB[j]][k] + X[(4*node)+k][PA[j]][k] for j in range(node))  <= 1 )
    
    #constraint (6)
    for k in range(vehicle):
        m.addConstr(quicksum(X[DB[i]][4*node+vehicle+k][k] + X[DA[i]][4*node+vehicle+k][k] for i in range(node)) <= 1)
    
    #constraint (7)
    for k in range(vehicle):
        for j in range(N):
            m.addConstr(quicksum(X[i][j][k] -X[j][i][k]  for i in range(N) ) == X[j][4*node+vehicle+k][k] - X[4*node+k][j][k])

    #constraint (8a)
    for i in range(N):
        for j in range(node):
            for k in range(vehicle):
                m.addGenConstrIndicator(X[i][PA[j]][k], True, T[i][k] +t[i][PA[j]] + w[PA[j]] -T[PA[j]][k] == 0)
                
    #constraint (8b)
    for i in range(N):
        for j in range(node):
            for k in range(vehicle):
                m.addGenConstrIndicator(X[i][PB[j]][k], True, T[i][k] +t[i][PB[j]] -T[PB[j]][k] == 0)
                m.addGenConstrIndicator(X[i][DB[j]][k], True, T[i][k] +t[i][DB[j]] -T[DB[j]][k] == 0)
                m.addGenConstrIndicator(X[i][DA[j]][k], True, T[i][k] +t[i][DA[j]] -T[DA[j]][k] == 0)
     
    #constraint (9a)
    for i in range(node): 
        for k in range(vehicle):
            m.addConstr(a[PA[i]]<= T[PA[i]][k] - dplus[PA[i]] )
            m.addConstr(T[PA[i]][k] - dplus[PA[i]] <= b[PA[i]])
            
    #constraint (9b)
    for i in range(node):
        for k in range(vehicle):
            m.addConstr(T[DB[i]][k] +dmin[DB[i]] -dplus[DB[i]] <= b[DB[i]])
            m.addConstr(a[DB[i]]<= T[DB[i]][k] +dmin[DB[i]] -dplus[DB[i]] )
    
    #constraint (10)
    for i in range(node):
        for k in range(vehicle):
            m.addConstr(T[PB[i]][k] + t[PB[i]][node+PB[i]] <= T[node+PB[i]][k])
            m.addConstr(T[PA[i]][k] + t[PA[i]][node+PA[i]] <= T[node+PA[i]][k])
            
    #constraint (11)
    for i in range(N):
        for j in range(N):
            for k in range(vehicle):
                m.addGenConstrIndicator(X[i][j][k], True, L[i][k] +l[j] - L[j][k] == 0)
                m.addGenConstrIndicator(X[4*node+k][j][k], True, L[4*node+k][k] +l[j] -L[j][k] == 0)
                
    #constraint (12)
    for i in range(node):
        for k in range(vehicle):
            m.addConstr(L[node+PB[i]][k] <= C[k] - l[node+PB[i]] )
            m.addConstr(L[node+PA[i]][k] <= C[k] - l[node+PA[i]] )
            
    #constraint (13)
    for k in range(vehicle):
        m.addConstr(L[4*node + k][k] == 0)
        
    #constraint 16 
    for i in range(node):
        m.addConstr(dmin[DB[i]] >= 0.0)
        m.addConstr(dplus[DB[i]] >= 0.0)
    for i in range(node): 
        m.addConstr(dplus[PA[i]] >= 0.0)
        m.addConstr(w[PA[i]] >= 0.0)
    
    m.addConstr(quicksum(quicksum(quicksum(c[i][j]*X[i][j][k] for j in range(4*node+ 2*vehicle))for i in range(4*node+ 2*vehicle))for k in range(vehicle)) == totalcost)
    
    m.optimize()
    m.write("constraint.lp")#check constraint via constraint.lp file!
    
    for v in m.getVars():#remove variable with value 0, cuz i dont need it
        if v.x != 0:
            print('%s %g' % (v.varName, v.x))
    
    print('Obj: %g' % m.objVal)
    

except GurobiError as e:
    print('Error code ' + str(e.errno) + ": " + str(e))

except AttributeError:
    print('Encountered an attribute error')
    

    
```
